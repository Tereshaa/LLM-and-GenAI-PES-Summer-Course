```python
from langchain.document_loaders.csv_loader import CSVLoader
import chardet

with open("/content/codebasics_faqs.csv","rb") as f:
  result=chardet.detect(f.read(10000))
  detected_encoding=result['encoding']
  print(detected_encoding)

loader=CSVLoader(file_path="/content/codebasics_faqs.csv",source_column="prompt",encoding=detected_encoding)
data=loader.load()

```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
model_name="BAAI/bge-large-en"
model_kwargs={"device": "cpu"}
encode_kwargs={"normalize_embeddings":True}
hf=HuggingFaceEmbeddings(model_name=model_name,model_kwargs=model_kwargs,encode_kwargs=encode_kwargs)
```

```python
e=hf.embed_query("what is ur refund policy")
print(e)
print(len(e))
```

```python
type(e)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

embeddings1=hf.embed_query("she is so lucky")
embeddings2=hf.embed_query("she is just fortunate")

embeddings1=np.array(embeddings1)
embeddings2=np.array(embeddings2)

embeddings1=embeddings1.reshape(1,-1)
embeddings2=embeddings2.reshape(1,-1)

similarity=cosine_similarity(embeddings1,embeddings2)
print(similarity)
```

```python
from langchain_community.embeddings import HuggingFaceBgeEmbeddings
from langchain.vectorstores import FAISS
embeddings=HuggingFaceBgeEmbeddings()
vector_db=FAISS.from_documents(documents=data,embedding=embeddings)
```

```python
retriever=vector_db.as_retriever()
r_docs=retriever.get_relevant_documents("for how long is this course valid")
r_docs
```

```python
from langchain.chains import RetrievalQA

chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retriever, input_key="query",return_score_documents=True)
```

```python
chain("do you guys provide internship and also do you offer emi payments")
```

```python
chain("do you guys provide javascript course?")
```

```python
from langchain.prompts import PromptTemplate

prompt_template = """Given the following context and a question, generate an answer based on this context only.
In the answer try to provide as much text as possible from "response" section in the source document context without making much changes.
If the answer is not found in the context, kindly state "I don't know." Don't try to make up an answer.

CONTEXT: {context}

QUESTION: {question}"""


PROMPT = PromptTemplate(
    template=prompt_template, input_variables=["context", "question"]
)
chain_type_kwargs = {"prompt":PROMPT}
```
